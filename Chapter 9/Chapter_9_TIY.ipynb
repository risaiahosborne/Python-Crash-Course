9-6. Ice Cream Stand

```python

class Restaurant:
    def __init__(self, restaurant_name, cuisine_type):
        self.restaurant_name = restaurant_name
        self.cuisine_type = cuisine_type
        self.number_served = 0

    def describe_restaurant(self):
        print(f"{self.restaurant_name} serves {self.cuisine_type} cuisine.")

    def open_restaurant(self):
        print(f"{self.restaurant_name} is now open!")
        
    def set_number_served(self, number):
        self.number_served = number
```

```python
class IceCreamStand(Restaurant):
    
    def __init__(self, restaurant_name, cuisine_type='Ice Cream'):
        super().__init__(restaurant_name, cuisine_type)
        self.flavors = []
        
    def display_flavors(self):
        print(f"{self.restaurant_name} offers the following flavors:")
        for flavor in self.flavors: #Loop through the flavors list and print each flavor
            print(f"- {flavor}")
```

```python
ice_cream_stand1 = IceCreamStand('Sweet Treats')
ice_cream_stand1.flavors = ['Vanilla', 'Chocolate', 'Strawberry']
ice_cream_stand1.display_flavors()
```

9-7. Admin

```python
class User:
    def __init__(self, first_name, last_name, email, password):
        self.first_name = first_name
        self.last_name = last_name
        self.email = email
        self.password = password
        self.login_attempts = 0
    
    def describe_user(self):
        print(f"User: {self.first_name} {self.last_name}")
        print(f"Email: {self.email}")
        
    def greet_user(self):
        print(f"Hello, {self.first_name}!")
        
    def increment_login_attempts(self):
        self.login_attempts += 1
        
    def reset_login_attempts(self):
        self.login_attempts = 0
        
class Admin(User):
    def __init__(self, first_name, last_name, email, password):
        super().__init__(first_name, last_name, email, password)
        self.privileges = [
            'can add post',
            'can delete post',
            'can ban user'
        ]
        self.privileges_instance = Privileges(self.privileges)  

class Privileges:
    def __init__(self, privileges):
        self.privileges = privileges
        
    def show_privileges(self):
        print("Admin privileges:")
        for privilege in self.privileges:
            print(f"- {privilege}")
            
```

```python
randy = User('Randy', 'Smith', 'randy@example.com', 'password123')
randy.describe_user()
randy.greet_user()

adminrandy = Admin('Randy', 'Smith', 'randy@example.com', 'password123')
adminrandy.privileges_instance.show_privileges() 
```

9-9. Battery Upgrade

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    
    def get_descriptive_name(self):
        long_name = f"{self.year} {self.make} {self.model}"
        return long_name.title()
    
    def read_odometer(self):
        print(f"This car has {self.odometer_reading} miles on it.")
        
    def update_odometer(self, mileage):
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")
            
    def increment_odometer(self, miles):
        self.odometer_reading += miles

class ElectricCar(Car):
    def __init__(self, make, model, year):
        super().__init__(make, model, year)
        self.battery = Battery()
        
class Battery:
    def __init__(self, battery_size=75):
        self.battery_size = battery_size
        
    def describe_battery(self):
        print(f"This car has a {self.battery_size}-kWh battery.")
        
    def get_range(self):
        if self.battery_size == 40:
            battery_range = 150
        elif self.battery_size == 65:
            battery_range = 225
        else:
            battery_range = 260
        print(f"This car can go about {battery_range} miles on a full charge.")
        
    def upgrade_battery(self):
        """ check the battery size and set the capacity to 65 if it isn't already """
        if self.battery_size != 65:
            self.battery_size = 65
            print("Battery has been changed to 65 kWh.")
```

```python
my_leaf = ElectricCar('nissan', 'leaf', 2024)
print(my_leaf.get_descriptive_name())
my_leaf.battery.describe_battery()
my_leaf.battery.get_range()
my_leaf.battery.upgrade_battery()
my_leaf.battery.get_range()
```
