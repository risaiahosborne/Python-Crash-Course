```python
alien_color = 'black'

if alien_color == 'blue':
    points = 5
elif alien_color == 'yellow':
    points = 10
elif alien_color == 'green':
    points = 15
elif alien_color == 'red':
    points = 20
else:
    points = 1

print(f'\n You earned {points} points for killing a {alien_color} alien!')
```

```python
age = 25

if age < 2:
    person = 'baby'
elif  2 <= age <4:
    person = 'toddler'
elif  4 <= age <12:
    person = 'kid'
elif  13 <= age <19:
    person = 'teenager'
elif  19 <= age <65:
    person = 'adult'
else:
    person = 'elder'

print(f'\n You are {age} years old that makes you an {person}!')
```
